```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = make_blobs(n_samples=1000, centers=5, n_features=2)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
linkage(x, method='centroid', metric='euclidean')
```

```python
clusters = fcluster(linkage(x, method='centroid', metric='euclidean'), 5, criterion="distance")
plt.scatter(x[:,0], x[:,1], c=clusters)
plt.show()
```

```python
for method in ['single', 'complete', 'average', 'weighted', 'centroid', 'median', 'ward']:
    distances = linkage(x, method=method, metric="euclidean")
    clusters = fcluster(distances, 3, criterion="distance") 
    plt.title('linkage: ' + method)
    plt.scatter(x[:,0], x[:,1], c=clusters, cmap='tab20b')
    plt.show()
```
