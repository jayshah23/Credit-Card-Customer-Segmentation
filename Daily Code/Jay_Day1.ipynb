filling null values 
1. mean - normally distributed data
2. median - data comprised of outliers
3. mode - data having more occurence of particular value

```python
df['MINIMUM_PAYMENTS'].value_counts().head()
```

for min_pay, we can use mean to fill null values

```python
# fetching null index
null_val = np.where(df['MINIMUM_PAYMENTS'].isnull())[0]
null_val
```

```python
print("Mean: "+str(df['MINIMUM_PAYMENTS'].mean()))
df['MINIMUM_PAYMENTS'] = df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean())
```
